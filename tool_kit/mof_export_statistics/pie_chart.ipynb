```python
import pandas as pd
import os
import plotly.express as px

from dotenv import load_dotenv

input_file ="mof_export_statistic.csv"


load_dotenv()


data_path = os.environ.get("PROCESSED_DATA_DIR")

fPath = os.path.join ( data_path, input_file ) 
df = pd.read_csv ( fPath )
df.head()
```

```python
cols_name = df.columns
cols_name
```

```python
#import plotly.io as pio
#pio.renderers.default = "browser" # 設定渲染器為瀏覽器

CLstr = "國家/地區別"
top10 = 10

#--------------------------------------------------------------------------
def pie_chart( colName , month , year , df ):
    df_sorted = ( 
                df [ ( df['月'] == month ) & ( df['年'] == year ) ]
                    .sort_values(by= colName , ascending=False)
                    .head(top10) 
                )
    df_selected = df_sorted [["年","月", CLstr, colName ]].copy()
    df_selected

    fig = px.pie (
        df_sorted,
        values=colName,
        names=CLstr,
        title=f"{colName} 前 10 國佔比 {year}年{month}月",
        width=800,
        height=600
    )

    fig.show()
#-----------------------------------------------------------------------------

year = 114
month = 1

colName = "總計" #"16.機械及電機設備" #"總計"
df = df[~df[ CLstr ].isin( ["北美洲","亞洲", "新南向18國(註1)", "東協10國", "中美洲", "歐洲"] ) ]

for year in range ( 114, 103, -1):
    for month in range ( 1, 13, 1 ):
        pie_chart( colName , month , year , df )

```
