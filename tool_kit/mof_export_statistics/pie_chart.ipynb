```python

#--------------------------------------------------------------------------
def pie_chart( colName , year , df ):

    order = sorted(df[CLstr].unique())  # 或自己定義 list
    fig = px.pie (
        df,
        values=colName,
        names=CLstr,
        color=CLstr,  # 用 CLstr 來決定顏色
        color_discrete_map=COLOR_MAP,  # 指定固定顏色
        title=f"{colName} 前 10 國佔比 {year}年",
        width=800,
        height=600,
        category_orders={CLstr: order}  # 這行控制順序
    )

    fig.show()
#-----------------------------------------------------------------------------

year = 114


colName = "總計" #"16.機械及電機設備" #"總計"
df = df[~df[ CLstr ].isin( ["北美洲","亞洲", "新南向18國(註1)", "東協10國", "中美洲", "歐洲"] ) ]

for year in range ( 114, 103, -1):
    df_yearly = df[df["年"] == year].copy()

    df_yearly = ( df_yearly.groupby( ["年", CLstr], as_index=False ) [ colName ].sum() )

    df_yearly["rank"] = df_yearly[colName].rank(method="dense", ascending=False)
    df_topN = df_yearly[df_yearly["rank"] <= topN].copy()
    df_topN = df_topN.sort_values ( [CLstr] )

    pie_chart( colName , year , df_topN )

```
